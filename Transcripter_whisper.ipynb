```python
# Copy of below cell, if you want word end times as well (for animation)

from faster_whisper import WhisperModel

model = WhisperModel("medium", compute_type="float32")  # You can also use "float32"

read_out_loud_story="new_audio.mp3"

segments, info = model.transcribe(read_out_loud_story, word_timestamps=True)

words = []
start_times = []
end_times = []

for segment in segments:
    for word in segment.words:
        print(f"{word.word}: {word.start:.2f}s – {word.end:.2f}s")
        words.append(word.word)
        start_times.append(round(word.start, 2))  # ✅ Fix here
        end_times.append(round(word.end, 2))


# Combine into a dict
data = {
    "words": words,
    "startTimes": start_times,
    "endTimes": end_times
}

# Save to JSON
import json
with open("words_and_start_and_end_times.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

# Download
from google.colab import files
files.download("words_and_start_and_end_times.json")
```
